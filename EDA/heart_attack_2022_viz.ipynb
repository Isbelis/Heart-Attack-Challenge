```python
normalized_pivot = round(pivoted_df.div( pivoted_df.sum( axis=1 ) , axis=0 ), 2)
normalized_pivot
```

```python
# Customize the plot function
def customize_plot(ax, title, xlabel, ylabel, title_size, label_size):
    ax.set_title(title, fontsize=title_size)
    ax.set_xlabel(xlabel, fontsize=label_size)
    ax.set_ylabel(ylabel, fontsize=label_size)

# Set figure size and style
plt.figure(figsize=(10, 8))
# https://stackoverflow.com/questions/74716259/the-seaborn-styles-shipped-by-matplotlib-are-deprecated-since-3-6
plt.style.use("seaborn-v0_8")  

# Plot the stacked bar chart
ax = pivoted_df.plot.bar(stacked=True, color=['tab:orange', 'tab:blue'], figsize=(10, 5))

# Customize x-tick rotation and plot titles
ax.tick_params(axis='x', rotation=0)
customize_plot(ax, 'Correlation b/w Heart Disease and Chronic Conditions', 'Chronic Conditions', 'People Count', 12, 10)

# Add labels for the stacked bars
for i, container in enumerate(ax.containers):
    if i == 0:  # Label for the 'No' category
        ax.bar_label(container, size=10, color='white', labels=pivoted_df['No'], label_type='center')
    else:  # Label for the 'Yes' category
        ax.bar_label(container, padding=-20, size=10, color='black', labels=pivoted_df['Yes'])
        ax.bar_label(container, padding=2, size=10, color='black', fmt='Total: %g')

# Display the plot
plt.show()
```

```python
Smoker_heart_attack_percentage = (SmokerData['HadHeartAttack'].value_counts(normalize=True) * 100).to_dict()
ECigarette_heart_attack_percentage = (ECigaretteData['HadHeartAttack'].value_counts(normalize=True) * 100).to_dict()
Alcohol_heart_attack_percentage = (AlcoholDrinkersData['HadHeartAttack'].value_counts(normalize=True) * 100).to_dict()
PhysicalActivities_heart_attack_percentage = (PhysicallyActiveData['HadHeartAttack'].value_counts(normalize=True) * 100).to_dict()


behavior_percentages = {
    'Smoking': Smoker_heart_attack_percentage.get('Yes'),
    'E-Cigarette Use': ECigarette_heart_attack_percentage.get('Yes'),
    'Alcohol Consumption': Alcohol_heart_attack_percentage.get('Yes'),
    'Physically Active': PhysicalActivities_heart_attack_percentage.get('Yes')
}
plt.pie(behavior_percentages.values(), labels=behavior_percentages.keys(), autopct='%1.1f%%', startangle=90,explode=[0,0.05,0.05,0.05])
plt.title('Prevalence of Health Behaviors among Heart Patients', weight = 'bold')
```

```python
pivot_table_mental_health = mental_health_cols.pivot_table(index='HadHeartAttack', columns='HadDepressiveDisorder', values='MentalHealthDays', aggfunc='median')
pivot_table_mental_health
```

```python
# Here values are aggregated by taking median of the values.
# For right highly skewed data, where most of the values are clustered towards the left with a long tail extending to the right, the best measure of central tendency is typically the median.
# The median is less affected by extreme values or outliers compared to the mean, making it a more robust measure for skewed distributions. It represents the middle value when the data is arranged in ascending or descending order. Since it is not influenced by extreme values, it provides a better representation of the typical value in the dataset for skewed distributions.
```

```python
# Loop for generating heatmap chart for all the risk factor columns

def senory_imp_cross_table_heat_map(df, column):
    # First melting the sensory impairment columns
    melted_df = df.melt(id_vars=[column], value_vars=['DeafOrHardOfHearing', 'BlindOrVisionDifficulty', 'DifficultyConcentrating'], var_name='SensoryImpairments', value_name='Status')
    
    # Select the rows with true status so that we can count only individuals with sensory impairments.
    melted_df = melted_df[melted_df['Status'] == 'Yes']
    
    # pivoting the melted df using cross tab function so that all the categories of sensory impairments becomes index and risk factors in columns 
    pivoted_df = round(pd.crosstab(index=melted_df['SensoryImpairments'], columns=melted_df[column], normalize=True),2)
    
    ax =sns.heatmap(data=pivoted_df, cmap='flare', linewidths=1, annot=True)

    customize_plot(ax, f'Heatmap: Normalized Correlation between Sensory Impairments and {column}',column, 'Sensory Impairment', 12,10)
    plt.show()
```

```python
senory_imp_cross_table_heat_map(sensory_imp_subset, 'SmokerStatus')
```

```python
# You can see in the chart that there are very less people with difficulty concentrating among smokers in comparison with non-smokers.
# It is due to the fact that smoking can improve concentration and attention. This is because a smoker's body needs a constant level of nicotine, and when these levels dip, it can cause withdrawal symptoms. As a result, the increased concentration after a cigarette is the body's way of thanking the smoker for satisfying its addiction.
# People with blindness and deafness, commonly donot smoke as there are very less people, around 1%-4%.
```
