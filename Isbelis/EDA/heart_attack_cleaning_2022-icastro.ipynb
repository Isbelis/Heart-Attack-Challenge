```python
# data science
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
## Load and merge dataset
```

```python
# we need to variable of coordinates
```

```python
# Load the CSV file "Origen"
df_geocode = pd.read_csv('./Resources/US_GeoCode.csv')
df_heart_2022 = pd.read_csv('./Resources/heart_2022_no_nans.csv')
```

```python
df_geocode.head(4)
```

```python
df_geocode.rename(columns={'Name': 'State'}, inplace=True)
df_geocode.info()
```

```python
df_heart_2022.head(4)
```

```python
# Merge the DataFrames based on a common key, such as 'shop_id' or 'location_id'
# Adjust the 'on' parameter based on the common key in your data
merged_df = pd.merge(df_heart_2022, df_geocode, on='State', how='left')
merged_df
```

```python
# Update latitude and longitude for Guam
merged_df.loc[merged_df['State'] == 'Guam', ['latitude', 'longitude']] = [13.444304, 144.793731]

# Update latitude and longitude for Virgin Islands
merged_df.loc[merged_df['State'] == 'Virgin Islands', ['latitude', 'longitude']] = [18.335765, -64.896335]

merged_df
```

```python
merged_df.describe()
```

```python
merged_df.info()
```

```python
# Save the merged DataFrame to a CSV file
merged_df.to_csv("merged_data_2022.csv", index=False)

## for my viz notebook 
```

```python
# Count the number of heart attacks by state
heart_attack_counts= merged_df['State'].value_counts()
heart_attack_counts
```

```python
#
```

```python
# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=heart_attack_counts.index, y=heart_attack_counts.values)
plt.title('Number of Heart Attacks by State')
plt.xlabel('State')
plt.ylabel('Number of Heart Attacks')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box Plot of BMI by Sex
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sex', y='BMI', data=merged_df)
plt.title('BMI Distribution by Sex')
plt.xlabel('Sex')
plt.ylabel('BMI')
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = merged_df[['PhysicalHealthDays', 'MentalHealthDays', 'SleepHours', 'BMI']].corr()
```

```python
# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Health Metrics')
plt.show()

```

```python
# Scatter Plot of Age vs. BMIYou can visualize the relationship between age and BMI.

plt.figure(figsize=(20,10))
sns.scatterplot(x='AgeCategory', y='BMI', data=merged_df, hue='HadHeartAttack', palette='deep')
plt.title('Scatter Plot of Age vs. BMI')
plt.xlabel('Age Category')
plt.ylabel('BMI')
plt.legend(title='Had Heart Attack')
plt.show()
```
