```python
# data science
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
heart_attack = pd.read_csv('heart_attack_2022_location.csv')
```

```python
heart_attack.head()
```

```python
# Count the number of heart attacks by state
heart_attack_counts = heart_attack['State'].value_counts()
heart_attack_counts
```

```python
# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=heart_attack_counts.index, y=heart_attack_counts.values, palette='viridis')
plt.title('Number of Heart Attacks by State')
plt.xlabel('State')
plt.ylabel('Number of Heart Attacks')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box Plot of BMI by Sex


plt.figure(figsize=(10, 6))
sns.boxplot(x='Sex', y='BMI', data=heart_attack, palette='Set2')
plt.title('BMI Distribution by Sex')
plt.xlabel('Sex')
plt.ylabel('BMI')
plt.show()
```

```python
# Heatmap of Correlation Between Health Metrics
# Calculate the correlation matrix
correlation_matrix = heart_attack[['PhysicalHealthDays', 'MentalHealthDays', 'SleepHours', 'BMI']].corr()
correlation_matrix
```

```python
# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Health Metrics')
plt.show()
```

```python
#  Scatter Plot of Age vs. BMI
plt.figure(figsize=(20, 10))
sns.scatterplot(x='AgeCategory', y='BMI', data=heart_attack, hue='HadHeartAttack', palette='deep')
plt.title('Scatter Plot of Age vs. BMI')
plt.xlabel('Age Category')
plt.ylabel('BMI')
plt.legend(title='Had Heart Attack')
plt.show()
```
